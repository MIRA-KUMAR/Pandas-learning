```python
urlretrieve('https://hub.jovian.ml/wp-content/uploads/2020/09/locations.csv', 'locations.csv')
```

```python
locations_df = pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df.loc[locations_df.location == 'Italy']
```

```python
covid_df['locations'] = 'Italy'
```

```python
covid_df
```

```python
locations_df = locations_df.rename(columns = {'location':'locations'})
```

```python
locations_df
```

```python
merged_data = covid_df.merge(locations_df, on="locations")
```

```python
merged_data
```

```python
merged_data['cases_per_million'] = (merged_data['new_cases'] *1e6) / merged_data['population']
```

```python
merged_data['deaths_per_million'] = (merged_data['new_deaths'] *1e6) / merged_data['population']
```

```python
merged_data['tests_per_million'] = (merged_data['new_tests'] *1e6) / merged_data['population']
```

```python
merged_data
```

```python
merged_data.columns
```

```python
results_df = merged_data[['date', 'new_cases', 'new_deaths', 'new_tests','locations' , 'population', 'life_expectancy',
       'hospital_beds_per_thousand', 'gdp_per_capita', 'cases_per_million',
       'deaths_per_million', 'tests_per_million']]
```

```python
results_df
```

```python
results_df.to_csv('results.csv', index=None)
```
